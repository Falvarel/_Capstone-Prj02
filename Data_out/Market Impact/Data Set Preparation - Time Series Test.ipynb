# INTRADAY ENERGY MARKET VALUE

## BIDS

```python
# From the recordtype column, select only records with the value 'BID':
bids = df[df['recordtype']=='BID']
bids.shape
```

```python
bids.loc[:, 'abs_acceptedprice'] = bids['acceptedprice'].abs()
bids.loc[:, 'abs_acceptedvolume'] = bids['acceptedvolume'].abs()
```

```python
bids.head()
```

```python
# Group by settlementDate and settlementPeriod values the abs of the sum of acceptedprice and acceptedvolume:
bids_sum = bids.groupby(['settlementDate','settlementPeriod',]).sum(['abs_acceptedprice', 'abs_acceptedvolume']).reset_index()
display(bids_sum.head())
```

```python
# Leave onty the columns if accepted price and accepted volume:
bids_sum = bids_sum[['settlementDate','settlementPeriod','abs_acceptedprice','abs_acceptedvolume']]

# Add a column with the week, month, and year of the settlement date:
bids_sum['settlementDate'] = pd.to_datetime(bids_sum['settlementDate'])
bids_sum['week'] = bids_sum['settlementDate'].apply(lambda x: x.isocalendar()[1])
bids_sum['week'] = bids_sum.apply(lambda row: row['week'] - 1 if row['settlementDate'].month == 1 and row['settlementDate'].day < 4 else row['week'], axis=1)
bids_sum['month'] = bids_sum['settlementDate'].dt.month
bids_sum['year'] = bids_sum['settlementDate'].dt.year   
bids_sum['day'] = bids_sum['settlementDate'].dt.day

# Display the result:
display(bids_sum.head())
```

```python
# Rename the acceptedprice and acceptedvolume columns:
bids_sum = bids_sum.rename(columns={'abs_acceptedprice':'sum_abs_acceptedprice', 'abs_acceptedvolume':'sum_abs_acceptedvolume'})
bids_sum.head()
```

### Accepted BIDS per week:

```python
# Create a table with the sum of accepted price and accepted volume for each week:
bids_sum_week = bids_sum.groupby(['year','month', 'week']).sum(['sum_abs_acceptedprice', 'sum_abs_acceptedvolume']).drop(['settlementPeriod','day'], axis =1).reset_index()
display(bids_sum_week.head(), len(bids_sum_week))
```

### Accepted BIDS per month:

```python
# Create a table with the sum of accepted price and accepted volume for each month:
bids_sum_month = bids_sum.groupby(['year', 'month']).sum(['sum_abs_acceptedprice', 'sum_abs_acceptedvolume']).drop(['settlementPeriod','day','week'], axis=1).reset_index()
display(bids_sum_month.head(), len(bids_sum_month))
```

### Accepted BIDS by year:

```python
# Create a table with the sum of accepted price and accepted volume for each year:
bids_sum_year = bids_sum.groupby(['year']).sum(['sum_abs_acceptedprice', 'sum_abs_acceptedvolume']).drop(['settlementPeriod', 'day', 'week', 'month'], axis =1).reset_index()
display(bids_sum_year.head(), len(bids_sum_year))
```

### Market volume of transactions in Pounds and MW:

```python
# Calculate the percentages for the new columns
bids_sum_year['market_volume_pounds'] = bids_sum_year['sum_abs_acceptedprice'] / 0.05
bids_sum_year['market_volume_MW'] = bids_sum_year['sum_abs_acceptedvolume'] / 0.05

# Adjust the values for the year 2023
bids_sum_year.loc[bids_sum_year['year'] == 2023, 'market_volume_pounds'] *= 2
bids_sum_year.loc[bids_sum_year['year'] == 2023, 'market_volume_MW'] *= 2

# Display the result:
display(bids_sum_year.head())
```

### Percentage (%) change in market value per pounds and MW:

```python
# Year over year growth:
bids_sum_year['market_value_yoy'] = bids_sum_year['market_volume_pounds'].pct_change()
bids_sum_year['market_liquidity_yoy'] = bids_sum_year['market_volume_MW'].pct_change()
bids_sum_year.head()
```

### Year to Date comparisson:

#### Selecting only the first 6 months:

```python
# Filter and keep only months from 1 to 6 for every year
filtered_bids_sum = bids_sum[(bids_sum['month'] >= 1) & (bids_sum['month'] <= 6)]

# Group by year and month, and sum the accepted price and accepted volume
bids_sum_month_ytd = filtered_bids_sum.groupby(['year', 'month']).sum(['sum_abs_acceptedprice', 'sum_abs_acceptedvolume']).drop(['settlementPeriod', 'day', 'week'], axis=1).reset_index()

# Display the result
display(bids_sum_month_ytd.head(25), len(bids_sum_month_ytd))
```

### Comparing YTD with for all 3 years:

```python
bids_sum_year_ytd = bids_sum_month_ytd.groupby(['year']).sum(['sum_abs_acceptedprice', 'sum_abs_acceptedvolume']).drop(['month'], axis=1).reset_index()
display(bids_sum_year_ytd.head(), len(bids_sum_year_ytd))
```

## BIDS:

```python
# From the recordtype column, select only records with the value 'BID':
df_bid = df[df['recordtype']=='BID']
df_bid.info()
```

```python
# Check for the maximum and minimum values of the 'acceptedprice' column:
df_bid['acceptedprice'].max(), df_bid['acceptedprice'].min()
```

### Outliers threshold:

```python
# Check for the amount of values higher and lower than the threshold value:
bid_outliersp_count = []
bid_outliersn_count = []
threshold = 3000 # Set the threshold value for the outliers

for i in df_bid['acceptedprice']:
    if (i <= -threshold):
        bid_outliersn_count.append(i)
    if (i >= threshold):
        bid_outliersp_count.append(i)

print('Number of outliers in the negative direction:', len(bid_outliersn_count),'proportion to total:', round(len(bid_outliersn_count)/len(df_bid)*100,6))
print('Number of outliers in the positive direction:', len(bid_outliersp_count), 'proportion to total:', round(len(bid_outliersp_count)/len(df_bid)*100,6))
```

```python
display(bid_outliersp_count)
```

```python
display(bid_outliersn_count)
```

### Remove the outliers:

```python
# Remove the outliers (values above threshold and below -threshold) in the acceptedprice column:
df_bid = df_bid[(df_bid['acceptedprice']<threshold) & (df_bid['acceptedprice']>-threshold)]
```

```python
# Check for the BMU Fuel Types:
df_bid['BMU Fuel Type'].unique()
```
