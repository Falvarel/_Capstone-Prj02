### - XGBOOST (BID MODEL - SP)

```python
file_paths = ['/content/drive/MyDrive/QEnergy/Results/XGBoost/XGBoost BID MODEL OPT. - SP (scores MAPE).csv',
              '/content/drive/MyDrive/QEnergy/Results/XGBoost/Archive/Old/XGBoost BID MODEL - SP (scores)+.csv',
              '/content/drive/MyDrive/QEnergy/Results/XGBoost/Archive/Old/XGBoost BID MODEL - SP (scores)+2.csv']
```

```python

print('XGBOOST (BID MODEL - SP)')

for x in file_paths:
  bidsp_scores = pd.read_csv(x)

  bidsp_rsme = bidsp_scores['rmse_test'].mean()
  bidsp_mae = bidsp_scores['mae_test'].mean()
  bidsp_meanreal = bidsp_scores['meanreal_test'].mean()

  bidsp_mape = (bidsp_scores['mae_test']/bidsp_scores['meanreal_test']).mean() * 100
  #bidsp_mape = (bidsp_mae/bidsp_meanreal) * 100

  print('_'*30)
  print(f'Avg. RSME: {bidsp_rsme:0.2f}')
  print(f'Avg. MAE: {bidsp_mae:0.2f}')
  print(f'Avg. REAL BID PRICE: {bidsp_meanreal:0.2f}')
  print(f'Avg. MAPE: {bidsp_mape:0.2f} %')
```
