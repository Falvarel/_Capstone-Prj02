```python
def split_30mins(input_df):

  out_df = input_df.copy()
  nr_rows = input_df.shape[0]


  for x in range(0,nr_rows):

    row_delta = input_df.loc[x,'row_delta']
    rows_delta = input_df.loc[x,'rows_delta']
    delta = row_delta + rows_delta

    st_time1 = 0
    end_time1 = 0


    if row_delta > 0.5:
      n = int(round(row_delta / 0.5, 0))

      out_df.loc[x,'CP_END_TIME'] = input_df.loc[x,'CP_ST_TIME'] + 0.5
      demand = round(input_df.loc[x,'FORECASTDEMAND'],2)

      for i in range(1,n):

        st_time1 = input_df.loc[x,'CP_ST_TIME'] + 0.5 * i
        end_time1 = input_df.loc[x,'CP_ST_TIME'] + 0.5 * (i + 1)

        newrow_values = {'SETTLEMENT_DATE':input_df.loc[x,'SETTLEMENT_DATE'], 'CP_ST_TIME':st_time1, 'CP_END_TIME':end_time1,	
                         'FORECASTDEMAND':demand,	'CARDINALPOINT':input_df.loc[x,'CARDINALPOINT'],	'CP_TYPE':input_df.loc[x,'CP_TYPE'], 
                         'row_delta':np.nan, 'rows_delta':np.nan}

        out_df.loc[len(out_df)] = newrow_values


    if rows_delta > 0.5 or (rows_delta >= 0.5 and row_delta > 0):

      if rows_delta >= 0.5 and row_delta == 0:
        n = int(round(rows_delta / 0.5, 0)) - 1
      else:
        n = int(round(rows_delta / 0.5, 0))

      demand_delta = input_df.loc[x+1,'FORECASTDEMAND'] - input_df.loc[x,'FORECASTDEMAND']

      for i in range(1,n+1):

        demand = round(input_df.loc[x,'FORECASTDEMAND'] + i/(n+1) * demand_delta,2)

        if st_time1 == 0 and end_time1 == 0:
          st_time2 = input_df.loc[x,'CP_ST_TIME'] + 0.5 * i
          end_time2 = input_df.loc[x,'CP_ST_TIME'] + 0.5 * (i + 1)
        else:
          st_time2 = st_time1 + 0.5 * i
          end_time2 = end_time1 + 0.5 * i


        newrow_values = {'SETTLEMENT_DATE':input_df.loc[x,'SETTLEMENT_DATE'], 'CP_ST_TIME':st_time2, 'CP_END_TIME':end_time2,	
                         'FORECASTDEMAND':demand,	'CARDINALPOINT':input_df.loc[x,'CARDINALPOINT'],	'CP_TYPE':input_df.loc[x,'CP_TYPE'], 
                         'row_delta':np.nan, 'rows_delta':np.nan}

        out_df.loc[len(out_df)] = newrow_values


  return out_df
```

```python
# CASES "CARDINAL" CATEGORIZATION:

# CASE 1:  1h  -> TAKE VALUES FROM 0.5h

# CASE 2:  (x-1) = x = (x+1)  ->  NO CHANGE
# CASE 3:  23.5h  -> TAKE VALUES FROM 24h
# CASE 4:  (x-1) != x = (x+1)  ->  NO CHANGE
# CASE 5:  (x-1) = x != (x+1)  ->  x CHANGE TO (x+1)
```

```python
def split_1hour(input_df):

  out_df = input_df.copy()
  nr_rows = input_df.shape[0]

  for x in range(0,nr_rows):

    st_time = out_df.loc[x,'CP_ST_TIME']
    #print(st_time)
    
    if st_time == 24:
      next
    elif st_time == 0.5:
      cardinal_x0 = 0
      cardinal_x1 = out_df.loc[x,'CARDINALPOINT']
      cardinal_x2 = out_df.loc[x+1,'CARDINALPOINT']
    else:
      cardinal_x0 = out_df.loc[x-1,'CARDINALPOINT']
      cardinal_x1 = out_df.loc[x,'CARDINALPOINT']
      cardinal_x2 = out_df.loc[x+1,'CARDINALPOINT']

    cptype_x1 = out_df.loc[x,'CP_TYPE']
    #print('continue')

    if (cardinal_x0 != cardinal_x1) & (cardinal_x1 != cardinal_x2) & (st_time != 24) & (str(st_time).replace('.', ',')[-2:] == ',5'):

      #print(st_time)
      out_df.loc[x+1,'CARDINALPOINT'] = cardinal_x1
      out_df.loc[x+1,'CP_TYPE'] = cptype_x1


  out_df['row_cat'] = out_df['CP_ST_TIME'].map(lambda x : False if str(x).replace('.', ',')[-2:] == ',5' else True)
  out_df['CP_ST_TIME'] = out_df['CP_ST_TIME'].map(lambda x : x+0.5 if str(x).replace('.', ',')[-2:] == ',5' else x)
  #print(out_df.head(5))


  out_dem = out_df.groupby(['SETTLEMENT_DATE','CP_ST_TIME'])['FORECASTDEMAND'].mean().round(2).reset_index(drop=True)
  #print(out_dem)


  out_df = out_df[out_df['row_cat']==True].reset_index(drop=True)
  #print(out_df.head(5))


  out_df['FORECASTDEMAND'] = out_dem
  #print(out_df.head(5))


  out_df.rename(columns={"CP_ST_TIME": "SETTLEMENT_PERIOD"},inplace=True)
  out_df = out_df[['SETTLEMENT_DATE','SETTLEMENT_PERIOD','FORECASTDEMAND','CARDINALPOINT','CP_TYPE']]


  return out_df
```
